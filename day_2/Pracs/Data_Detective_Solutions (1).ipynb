**Section 2: Preprocessing**

---



2.1 Check for any missing data (NaNs/nulls) and quantify the amount of missing data

```python
df.isnull().sum()
```

2.2 Handle the missing values using an appropriate method.

```python
df['age'] = df['age'].fillna(df['age'].median()) #numerical column
```

```python
#categorical columns
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])

df['embark_town'] = df['embark_town'].fillna(df['embark_town'].mode()[0])
```

```python
df = df.drop(columns=['deck'])
```

2.3 Check for duplicate rows and remove them if any exist.

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

2.4 Detect outliers using the IQR method.

```python
Q1 = df['age'].quantile(0.25)
Q3 = df['age'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['age'] < lower) | (df['age'] > upper)]

print(outliers)
```

```python
print(len(outliers))
```

2.5 Encode the categorical variables.

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

df['sex'] = encoder.fit_transform(df['sex'])
df['embarked'] = encoder.fit_transform(df['embarked'])
df['class'] = encoder.fit_transform(df['class'])
df['who'] = encoder.fit_transform(df['who'])
df['adult_male'] = encoder.fit_transform(df['adult_male'])
df['alive'] = encoder.fit_transform(df['alive'])
df['alone'] = encoder.fit_transform(df['alone'])
```

2.5 Create a new feature called family_size

```python
df['family_size'] = df['sibsp'] + df['parch'] + 1
```

```python
df['is_child'] = (df['age'] < 18).astype(int)
```

2.6 Normalize the numerical features.

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

columns = ['age','fare','family_size']

df[columns] = scaler.fit_transform(df[columns])
```

2.7 Randomize the dataset.

```python
df = df.sample(frac=1, random_state=42).reset_index(drop=True)
```

2.8 Produce a correlation heatmap.

```python
corr = df.corr(numeric_only=True)

corr

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))

sns.heatmap(corr,
            annot=True,
            cmap="coolwarm")

plt.show()
```

Save the cleaned dataset as a CSV file.

```python
df.to_csv("titanic_cleaned.csv", index=False)
```
