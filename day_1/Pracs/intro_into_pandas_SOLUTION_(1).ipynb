##### 3.1 No headers


---



```python
# imports
import pandas as pd

# get the file path
file_path = 'no_headers.xlsx'

# Read the Excel file into a DataFrame
df = pd.read_excel(file_path)

df.head()
```

```python
# You can specify to pandas that there are no headers:
df = pd.read_excel(file_path, header=None)

# It then defaults to numbering the columns
df.head()
```

```python
# You can then create your 'own' headers
# specify your header names
column_names = ['Full_Name', 'Province', 'Age', 'Income', 'Employment_Status',
                'Credit_Score', 'Home_Owner', 'Has_Debt', 'Loan_Approved']

df = pd.read_excel(file_path, header=None, names=column_names)

df.head()
```

# Part 5: Modifying DataFrames


---



### 5.1 Replacing

```python
# imports
import pandas as pd
import numpy as np

# Read the Excel file into a DataFrame
df = pd.read_excel('dirty_detailed.xlsx')
df.head(10)
```

```python
# Let us replace -- on numbered columns with nan
numeric_cols = ['Attendance', 'Temperature (°C)', 'Shots on Target', 'Yellow Cards', 'Win Probability']
df[numeric_cols] = df[numeric_cols].replace('--', np.nan)
df
```

```python
# Replacing in specific columns.
df['Win Probability'] = df['Win Probability'].replace('--', np.nan)
df
```

### 5.2 Removing

```python
# imports
import pandas as pd

# Read the Excel file into a DataFrame
df = pd.read_excel('dirty_detailed.xlsx')
df
```

```python
# We want to replace rows with Team names  as --
df['Team Name'] != '--'
```

```python
# If we include the original dataframe, we will only get instances where the result is true
df = df[df['Team Name'] != '--']
df
```

```python
# We the need to reset the index
df = df.reset_index()
df
```

### Now let us move to removing columns

```python
# We want to remove League and index
# Using Drop
df = df.drop(columns=['League', 'index'])
df
```

```python
# We want to keep only Team Name, Opponent and Win Probability
# Using Filter
# axis=1 for colunms
df_filtered = df.filter(items=['Team Name', 'Opponent', 'Win Probability'], axis=1)
df_filtered
```

#### 5.3 Dropping Duplicates

```python
import pandas as pd

# Read the file into a DataFrame (duplicates sheet)
df = pd.read_excel('dirty_data.xlsx', sheet_name='duplicates')
df
```

```python
# Removing Duplicates
# Have an option of completely dropping by setting keep = False
# You can keep the last entry by specifyin keep = 'last'
df_no_dups = df.drop_duplicates()
df_no_dups
```

```python
# Reset the index
df_no_dups = df_no_dups.reset_index(drop=True)
df_no_dups
```
